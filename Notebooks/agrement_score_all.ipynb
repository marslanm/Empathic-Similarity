```python
import pandas as pd
import numpy as np
import krippendorff
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import cohen_kappa_score

def continuous2discrete(numbers):
    data = []
    for n in numbers:
        if n - int(n) < 0.5:
            data.append(int(n))
        else:
            data.append(int(n) + 1)
    return data

def cal_IAA(data):
    results = {}
    for aspect in list(data.keys()):
        annotations = data[aspect]
        num_annotators = len(annotations)

        # Calculate Pearson and Spearman correlation coefficients
        pearson_corrs = []
        spearman_corrs = []
        cohen_kappa = []
        # # for i in range(0,1):
        # #        j=1
        # for i in range(2,6):
        #   for j in range(i + 1, 6):
        for i in range(num_annotators):
          for j in range(i + 1, num_annotators):
            pearson_corr, _ = pearsonr(annotations[i], annotations[j])
            spearman_corr, _ = spearmanr(annotations[i], annotations[j])
            kappa = cohen_kappa_score(continuous2discrete(annotations[i]), continuous2discrete(annotations[j]))
            pearson_corrs.append(pearson_corr)
            spearman_corrs.append(spearman_corr)
            cohen_kappa.append(kappa)

        # Average Pearson and Spearman correlation coefficients
        average_pearson_corr = np.mean(pearson_corrs)
        average_spearman_corr = np.mean(spearman_corrs)
        average_cohen_kappa = np.mean(cohen_kappa)

        # Calculate Krippendorff's Alpha\A
        # larger Krippendorffâ€™s Alpha means higher agreement among the annotators, ranges from -1 to 1, where:
        # 1 indicates perfect agreement.
        # 0 indicates no agreement better than chance.
        # Negative values indicate systematic disagreement.
        alpha = krippendorff.alpha(reliability_data=np.array(annotations), level_of_measurement='interval')


        results[aspect] = {
            "Average Pearson Correlation": average_pearson_corr,
            "Average Spearman Correlation": average_spearman_corr,
            # a higher alpha value signifies that the annotators are more consistent in their labeling,
            # whereas a lower alpha value indicates greater disagreement.
            "Krippendorff's Alpha": alpha,
            "Average Cohen Kappa": average_cohen_kappa,
        }

    return pd.DataFrame(results)
```

```python
results = cal_IAA(data = df)
print(results.to_latex(float_format="%.3f"))
results
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import cohen_kappa_score
import krippendorff

# Load the DataFrame (assuming df is your DataFrame)
# df = pd.read_csv('path_to_your_file.csv')

# Compute average empathy scores
empathy_scores = np.mean(
    [
        df.iloc[:, 2].values,
        df.iloc[:, 5].values,
        df.iloc[:, 8].values,
        df.iloc[:, 11].values
    ],
    axis=0
)

# Extract GPT scores
gpt_scores = df.iloc[:, 12].values

# Compute average Pearson correlation
average_pearson = pearsonr(empathy_scores, gpt_scores)[0]

# Compute average Spearman correlation
average_spearman = spearmanr(empathy_scores, gpt_scores)[0]

# Prepare data for Krippendorff's Alpha
krippendorff_data = np.vstack([empathy_scores, gpt_scores])

# Compute Krippendorff's Alpha
krippendorff_alpha = krippendorff.alpha(reliability_data=krippendorff_data, level_of_measurement='interval')

# Compute average Cohen's Kappa (we need to bin the scores to discrete values first)
empathy_scores_binned = np.digitize(empathy_scores, bins=np.linspace(1, 4, 4))
gpt_scores_binned = np.digitize(gpt_scores, bins=np.linspace(1, 4, 4))
average_cohen_kappa = cohen_kappa_score(empathy_scores_binned, gpt_scores_binned)

# Print the results
print(f"Average Pearson Correlation: {average_pearson}")
print(f"Average Spearman Correlation: {average_spearman}")
print(f"Krippendorff's Alpha: {krippendorff_alpha}")
print(f"Average Cohen's Kappa: {average_cohen_kappa}")
```
